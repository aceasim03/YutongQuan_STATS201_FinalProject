## Question 4: 




There have been several changes to Ethereum before and after the merge. To name a few, there is Ethereum 2.0, sharding, and staking. 

**Ethereum 2.0**: The merge refers to moving from Ethereum 1.0 to Ethereum 2.0, which uses a Proof-of-Stake (PoS) mechanism instead of the current Proof-of-Work (PoW) mechanism (consensys 2020). This is expected to increase security and scalability and reduce consumption. 

**Sharding**: Sharding is a new way to divide the network into smaller pieces, and segments, to increase speed and capacity (ethereum.org 2023). Now, the network can process more transactions simultaneously, which will certainly improve efficiency and scalability. 

**Staking**: As opposed to miners being responsible for verifying transactions and adding blocks to the blockchain, Ethereum 2.0 will require validators to be responsible (ethereum.org 2023). Stakers will earn rewards for verifying transactions and adding blocks, which gives them the incentive to participate in the network. This is expected to increase network stability and security.

| Feature | Before Merge | After Merge | 
| -------- |--------------|--------|
| Consensus Mechanism	| Proof-of-Work (PoW) | Proof-of-Stake (PoS) | 
| Sharding | Not available | Available |
| Responsibility of Transaction Validation	 | Miners | Validators
